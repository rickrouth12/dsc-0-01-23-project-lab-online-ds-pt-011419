```python
import requests, string
macbeth = requests.get('http://www.gutenberg.org/cache/epub/2264/pg2264.txt').text
```

```python
str_list = macbeth.split()  # list of words
updated_list = macbeth.translate(string.punctuation) # omit punctuation marks
unique_words = set(str_list) # create set of unique words in list
word_counts = {}
for words in unique_words : 
    word_counts[words] = {words, updated_list.count(words)}
print(word_counts.values())
```
